```python
missing_value=['N/a','na','NaN']
allrounders_df=pd.read_csv('All_rounders.csv',na_values=missing_value)
allrounders_df.isnull().sum()
```

```python
allrounders_df.rename(columns={'allround//0':'id','allround/Player':'Player','allround/Country':'country','allround/Inns':'Innings',
                              'allround/Runs':'Runs','allround/Avg':'Batting_avg','allround/Wkts':'wickets','allround/Avg_Bowl':'Bowling_avg',
                              'allround/Ca':'Catches','allround/Mat':'mat','allround/Overs':'overs','allround/Rating':'rating'},inplace=True)
```

```python
allrounders_df
```

```python
allrounders_df.drop(['mat','overs','rating'], axis=1 , inplace=True )
allrounders_df
```

```python
allrounders_df.dropna(inplace=True)
```

```python
allrounders_df = allrounders_df.astype({"wickets":'int', "Catches":'int'})
```
