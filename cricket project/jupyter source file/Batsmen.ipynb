```python
missing_value=['N/a','na','NaN']
batsmen_df=pd.read_csv('raw_batsmen_data.csv',na_values=missing_value)
batsmen_df.isnull().sum()
```

```python
batsmen_df.rename(columns={'stats/50':'50s','stats/Player':'Player','stats/Inns':'Innings','stats/NO':'Not_outs',
                              'stats/Runs':'Runs','stats/HS':'Highest score','stats/Ave':'batting_average','stats/SR':'Strike_rate'
                              },inplace=True)
```

```python
batsmen_df
```

```python
def swap_columns(df, col1, col2):
    col_list = list(df.columns)
    x, y = col_list.index(col1), col_list.index(col2)
    col_list[y], col_list[x] = col_list[x], col_list[y]
    df = df[col_list]
    return df
```

```python
swap_columns(batsmen_df,'50s','Player')
```
